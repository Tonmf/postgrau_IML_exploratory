```
SL = dfI_num.loc[dfI_num['SepalLengthCm'].isna()==True].shape[0]
SW = dfI_num.loc[dfI_num['SepalWidthCm'].isna()==True].shape[0]
PL = dfI_num.loc[dfI_num['PetalLengthCm'].isna()==True].shape[0]
PW = dfI_num.loc[dfI_num['PetalWidthCm'].isna()==True].shape[0]
print('Hay {} espacios en blanco de SL'.format(SL))
print('Hay {} espacios en blanco de SW'.format(SW))
print('Hay {} espacios en blanco de PL'.format(PL))
print('Hay {} espacios en blanco de PW'.format(PW))
```

Ejercicio 4


```
dfI.groupby(by=['Species']).mean()
```

```
#Numero total de species
NIset=dfI.loc[dfI['Species']=='Iris-setosa'].shape[0]
NIver=dfI.loc[dfI['Species']=='Iris-versicolor'].shape[0]
NIvir=dfI.loc[dfI['Species']=='Iris-virginica'].shape[0]
print('Hay {} flores de Iris-setosa, {} de Iris-versicolor y {} de Iris-virginica'.format(NIset,NIver,NIvir))
```

```
NSLmax = dfI.loc[dfI['SepalLengthCm']>=5.8].shape[0]
print('Hay {} con el Sepal más largo de 5.8cm'.format(NSLmax))
```

Ejercicio 5

```
import seaborn as sns
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,6))
ax = sns.lineplot(x="SepalLengthCm", y="SepalWidthCm",hue='Species', data=dfI)
plt.show()
```

```
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(211)
ax = sns.barplot(x="Species", y="PetalWidthCm", data=dfI)
ax1 = fig.add_subplot(212)
ax1 = sns.barplot(x="Species", y="PetalLengthCm", data=dfI)
plt.show()
```
